***Fourier Transforms***

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,10*np.pi,100)
x = np.sin(2*np.pi*t) + np.sin(4*np.pi*t) + 0.1*np.random.randn(len(t))
plt.plot(t,x)
```

```python
from scipy.fft import fft, fftfreq
N = len(x)
y = fft(x)[:N//2]
```

```python
plt.plot(np.abs(y))
```

```python
f = fftfreq(N,np.diff(t)[0])[:N//2]
```

```python
plt.plot(f,np.abs(y))
```
