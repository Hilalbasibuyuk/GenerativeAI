```python
import matplotlib.pyplot as plt
import numpy as np
import PIL as pillow
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
flowers_url = "https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz"
```

```python
import pathlib
flowers_data = tf.keras.utils.get_file("flower_photos", origin = flowers_url, untar = True)
flowers_data = pathlib.Path(flowers_data)
```

```python
flowers_data
```

```python
import glob
all_sunflowers = list(flowers_data.glob("sunflowers/*"))
```

```python
print(all_sunflowers[1])
pillow.Image.open(str(all_sunflowers[1]))
```

```python
height, width = 180, 180
```

```python
training_batch_size = 32
```

```python
train_set = tf.keras.preprocessing.image_dataset_from_directory(flowers_data,
                                                               validation_split = .2,
                                                               subset = "training",
                                                                seed = 123,
                                                                image_size = (height,width),
                                                                batch_size =training_batch_size)
```

```python
image_cat = train_set.class_names
image_cat
```

```python
validation_set = tf.keras.preprocessing.image_dataset_from_directory(
flowers_data,
validation_split=0.2,
subset="validation",
seed=123,
image_size=(height, width),
batch_size=training_batch_size)
```

```python
model = Sequential()
imported_model = tf.keras.applications.ResNet50(include_top = False,
                                               input_shape = (180,180,3),
                                               pooling = "avg",
                                               classes = 5,
                                               weights = "imagenet")
```

```python
for layer in imported_model.layers:
    layer.trainable = False
```

```python
from tensorflow.python.keras.layers import Dense, Flatten
model.add(imported_model)
model.add(Flatten())
model.add(Dense(512,activation = "relu"))
model.add(Dense(5, activation = "softmax"))
```

```python
model.summary()
```

```python
from tensorflow.keras.optimizers import Adam
model.compile(optimizer = Adam(lr = 0.001), loss = "sparse_categorical_crossentropy", metrics = ["accuracy"])
```

```python
history = model.fit(train_set,validation_data = validation_set, epochs = 10)
```

```python
import cv2
```

```python
image=cv2.imread(str(all_sunflowers[1]))
image_resized= cv2.resize(image, (height, width))
image=np.expand_dims(image_resized,axis=0)
print(image.shape)
```

```python
model_pred=model.predict(image)
```

```python
redicted_class=image_cat[np.argmax(model_pred)]
print("The predicted category is",redicted_class)
```
