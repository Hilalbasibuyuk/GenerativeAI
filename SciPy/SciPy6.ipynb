```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,10)
y = x**2 * np.sin(x)
plt.scatter(x, y)
```

Bu değerler arasında ne var bunu öğrenmek istiyoruz

```python
from scipy.interpolate import interp1d # İnterpolasyon ile ara değer buluruz eksik veriler bile olsa (ortalama)
```

```python
f = interp1d(x, y, kind = 'linear')
x_yogunluk = np.linspace(0,10,100)
y_yogunluk = f(x_yogunluk)
plt.plot(x_yogunluk, y_yogunluk)
```

```python
f = interp1d(x, y, kind = 'cubic')
x_yogunluk = np.linspace(0,10,100)
y_yogunluk = f(x_yogunluk)
plt.plot(x_yogunluk, y_yogunluk)
```
