```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data = np.linspace(0,10,10)
y_data = 3*x_data**2 + 2
plt.scatter(x_data,y_data)
```

```python
from scipy.optimize import curve_fit
def func(x, a, b):
    return a*x**2+b
popt, pcov = curve_fit(func, x_data, y_data, p0 = (1,1))
```

```python
popt
```

```python
plt.scatter(x_data, y_data, label='Veri Noktaları')
x_fit = np.linspace(0, 10, 100)
y_fit = func(x_fit, *popt) # *popt, popt'un her bir elemanını ayrı parametreler olarak alır
plt.plot(x_fit, y_fit, 'r-', label='Eğri Uyumu')
plt.legend()
plt.show()
```
