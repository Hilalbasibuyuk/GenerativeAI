```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

Differential Equations

```python
def dvdt(v, t):
    return 3*v**2-5
v0 =0
```

```python
t = np.linspace(0,1,100)
solution = odeint(dvdt, v0, t)
```

```python
solution.T # T daha sütunlarla da almak için
```

```python
plt.plot(t,solution.T[0])
```

Coupled first order ODEs

```python
def dSdx(S,x):
    y1,y2 = S
    return [y1 + y2**2 +3*x, 3*y1 +y2**3-np.cos(x)] # Rastgele fonksiyonlar
y1_0 = 0
y2_0 = 0
S_0 = (y1_0,y2_0)
```

```python
x = np.linspace(0,1,100)
solution = odeint(dSdx, S_0, x)
```

```python
solution.T
```

```python
y1 = solution.T[0]
y2 = solution.T[1]
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
```

Second Order ODEs

```python
def dSdt(S,t):
    theta,omega = S
    return [omega,
           np.sin(theta)]
theta0 = np.pi/4
omega0 = 0
S0 = (theta0,omega0)
```

```python
t = np.linspace(0,20,100)
solution = odeint(dSdt,S0,t)
```

```python
theta,omega = solution.T
```

```python
plt.plot(t,theta)
```
