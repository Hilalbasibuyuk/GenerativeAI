```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(0)
```

```python
data = {
    'A':np.random.randint(0,100,10),
    'B':np.random.randint(0,100,10),
    'C':np.random.randint(0,100.10)
}
```

```python
df = pd.DataFrame(data)
df
```

```python
df.describe()
```

```python
print("Sadece 2. ve 3. satırlar") #Satır seçme
df.iloc[1:3]
```

```python
print("Sadece A ve B sütunları") #Sütun seçme
df[["A","B"]]
```

```python
df['C'] = df['C'] * 2 #Sütunda işlem yapma
df['C']
```

```python
df['D'] = df['A'] + df['B'] # Yeni sütun ekleme
df
```

```python
print("Sadece 50'den büyük değerler A sütunu için")
df[df['A']>50]
```

```python
df.groupby('B').sum() # B sütununa göre gruplama ve toplama
```

```python
df.iloc[0,0] = np.nan # NAN değerler ekleme işlemi
df.iloc[2,1] = np.nan
df
```

```python
df = df.fillna(df.mean()) # nan değerleri ortalama değerler ile doldurma
df
```

```python
df = df.rename(columns ={ "A":"X","B":"Y","C":"Z","D":"W"})
df
```

```python
df.to_csv("veri.csv",index = False) # Veri dosyaya yazdırılır
```
