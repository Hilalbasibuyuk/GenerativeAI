```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
from keras.applications.nasnet import NASNetLarge
```

```python
train_path = "/content/drive/MyDrive/Colab Notebooks/Data/train/"
test_path = "/content/drive/MyDrive/Colab Notebooks/Data/test/"
```

```python
train_data = ImageDataGenerator(rescale= 1./255, shear_range=0.2, rotation_range = 15, zoom_range = 0.1, horizontal_flip=True)
test_data = ImageDataGenerator(rescale=1./255)
```

```python
train_datagen = train_data.flow_from_directory(train_path,
                                               target_size = (224,224),
                                               batch_size = 32,
                                               color_mode = "rgb",
                                               class_mode = "categorical")
```

```python
test_datagen = test_data.flow_from_directory(test_path,
                                             target_size = (224,224),
                                             batch_size = 32,
                                             color_mode = "rgb",
                                             class_mode  = "categorical")
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, GlobalAveragePooling2D

base_model = NASNetLarge(weights = "imagenet", include_top=False, input_shape = (224,224,3))

model = Sequential()
model.add(base_model)
model.add(GlobalAveragePooling2D())
model.add(Dense(64, activation= "relu"))
model.add(Dropout(0.5))
model.add(Dense(32, activation="relu"))
model.add(Dense(4, activation = "softmax"))
```

```python
model.compile(optimizer = "adam", loss = "categorical_crossentropy", metrics = ["accuracy"])
```

```python
import keras
es = keras.callbacks.EarlyStopping(monitor='val_accuracy', mode='auto', restore_best_weights=True, verbose=1, patience=7)
```

```python
history = model.fit(train_datagen, validation_data= test_datagen, batch_size= 64, epochs = 30, callbacks= [es], verbose = 1)
```
