```python
from scipy.special import legendre
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,1,100)
```

```python
plt.plot(x, legendre(6)(x)) # Legendre polinomunu hesaplar
plt.show()
```

```python
legendre(6)(x)
```

```python
from scipy.special import jv
x = np.linspace(0,10,100)
plt.plot(x, jv(3,x))
plt.show()
```
