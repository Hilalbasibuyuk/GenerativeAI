```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
planets = sns.load_dataset("planets")
```

```python
tips
```

```python
sns.scatterplot(x = "tip", y = "total_bill", data = tips)
```

```python
sns.scatterplot(x = "tip", y = "total_bill", data = tips, hue = "day")
```

```python
sns.scatterplot(x = "tip", y = "total_bill", data = tips, hue = "day",size = "size")
```

```python
sns.scatterplot(x = "tip", y = "total_bill", data = tips, hue = "day",size = "size",palette = "YlGnBu")
```

```python
sns.histplot(tips["tip"], kde = True, bins = 20) #kde grafikteki Ã§izgidir. KaldÄ±rÄ±rsak Ã§izgi de gider
```

```python
sns.displot(tips["tip"], kde = True, bins = 20)
```

```python
sns.distplot(tips["tip"], kde = True, bins = 20)
```

```python
sns.barplot(x="sex",y="tip",data = tips,palette = "YlGnBu")
```

```python
sns.boxplot(x="day",y="tip", data = tips, hue="sex",palette = "YlGnBu")
```

```python
sns.stripplot(x="day", y = "tip",data = tips, hue = "sex",palette = "YlGnBu")
```

```python
sns.stripplot(x="day", y = "tip",data = tips, hue = "sex",palette = "YlGnBu",dodge = True) #Dodge ile ayrÄ±lÄ±r
```

```python
sns.jointplot(x="tip", y="total_bill", data = tips)
```

```python
sns.jointplot(x = "tip", y = "total_bill", data = tips, kind = "reg")
```

```python
sns.jointplot(x = "tip", y = "total_bill", data = tips, kind = "kde")
```

```python
sns.jointplot(x = "tip", y = "total_bill", data = tips, kind = "kde", fill = True)
```

```python
sns.jointplot(x = "tip", y = "total_bill", data = tips, kind = "kde",fill = True, cmap = "YlGnBu")
```

```python
sns.jointplot(x = "tip", y = "total_bill", data = tips, kind = "hex",cmap = "YlGnBu")
```
