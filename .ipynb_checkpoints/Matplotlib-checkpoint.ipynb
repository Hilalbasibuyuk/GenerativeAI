```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1,10,100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
fig,ax = plt.subplots(nrows = 2,ncols =1, figsize=(10,8))
ax[0].plot(x,y1,label='sinx',color="blue",linestyle='-.')
ax[1].plot(x,y2,label="cosx",color = "red",linestyle='--')
ax[0].set_title("Sinüs fonksiyonu")
ax[1].set_title("Cosinüs fonksiyonu")
fig.suptitle('Sinüs ve kosinüs fonksiyonları',fontsize=16)
for i in range(2):
    ax[i].set_xlabel("x")
    ax[i].set_ylabel("y")
    ax[i].grid(True)
    ax[i].legend()
plt.tight_layout()
plt.show() #Hepsini aynı yerde yazmayınca grafik çizdirilmiyor
```
