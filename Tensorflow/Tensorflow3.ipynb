```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("is_memnuniyeti.csv")
data.head(7)
```

```python
x = pd.get_dummies(data.drop(["left"],axis=1))
x = x.astype(int)
y = data["left"].apply(lambda x:1 if x ==1 else 0)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size =.2)
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
```

```python
model = Sequential()
model.add(Dense(units=32, activation = "relu", input_dim = len(x_train.columns)))
model.add(Dense(units=64, activation = "relu"))
model.add(Dense(units =1, activation = "sigmoid"))
```

```python
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics='accuracy')
```

```python
model.fit(x_train,y_train,epochs = 200, batch_size = 32)
```

```python
y_sonuc = model.predict(x_test)
y_sonuc = [0 if val<0.5 else 1 for val in y_sonuc]
```

```python
y_sonuc
```

```python
accuracy_score(y_test,y_sonuc)
```
